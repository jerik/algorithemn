```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('pokemon.csv', index_col=0)
```

```python
df
```

```python
sns.lmplot(x='Attack',y='Defense', data=df)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df,
          fit_reg=False)
```

```python
sns.lmplot(x='Attack', y='Defense', data=df, 
          fit_reg=False, hue='Stage')
```

```python
sns.lmplot(x='Attack', y='Defense', data=df, 
          fit_reg=False, hue='Legendary')
```

```python
sns.lmplot(x='Attack', y='Defense', data=df, 
          fit_reg=False, hue='Legendary')
# Tweak using matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
# Boxplot
sns.boxplot(data=df)
```

```python
# Pre-format Dataframe
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

sns.boxplot(data=stats_df)
```

```python
sns.set_style('whitegrid')

# Set the size of the figure
# https://stackoverflow.com/a/57514771/1933185
plt.figure(figsize=(14,8))
sns.violinplot(x='Type 1', y='Attack', data=df)
```

```python
plt.figure(figsize=(14,8))
sns.swarmplot(x='Type 1', y='Attack', data=df)
```

```python
plt.figure(figsize=(10,6))

sns.violinplot(x='Type 1',
               y='Attack',
               data=df,
               inner=None # Remove the bars inside the violins
              )

sns.swarmplot(x='Type 1',
              y='Attack',
              data=df,
              color='k', # Make points black
              alpha=0.7 # And slightly transparent
             )

plt.title('Attack by Type')
```

```python
stats_df
```

```python
# Melt dataframe
melted_df=pd.melt(stats_df,
                   id_vars=['Name', 'Type 1', 'Type 2'], # variables to keep
                   var_name='stat' # Name of the melted variable
                  )
```

```python
melted_df
```

```python
print(stats_df.shape)

print(melted_df.shape)
```

```python
plt.figure(figsize=(14,8))
sns.swarmplot(x='stat', y='value', data=melted_df, hue='Type 1')
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x='stat',
              y='value',
              data=melted_df,
              hue='Type 1',
              dodge=True)

plt.ylim(0,260)
plt.legend(bbox_to_anchor=(1,1), loc=2)
```

```python
# Calculate correlations
corr=stats_df.corr()

sns.heatmap(corr)
```

```python
# Distribution plot (a.k.a Histogram)
sns.distplot(df.Attack)
```

```python
sns.distplot(df.Defense)
```

```python
sns.countplot(x='Type 1', data=df)

plt.xticks(rotation=-45)
```

```python
# Factorplot 
# is renamed to catplot, default kind = point is changed to kind = strip in catplot
g = sns.catplot(x='Type 1',
                  y='Attack',
                  data=df,
                  hue='Stage', # Color by stage
                  col='Stage', # Separate by stage
                  kind='swarm' # Swarmplot
                 )
# Rotate x-axis lables
g.set_xticklabels(rotation=-45)
```

```python

# Desnsity plot
plt.figure(figsize=(10,6))
sns.kdeplot(df.Attack, df.Defense)
# Combining with scatterplot seems not to work :(
sns.lmplot(y='Defense', x='Attack', data=df, fit_reg=False)
```

```python
# Joint Distribution plot
sns.jointplot(x='Attack', y='Defense', data=df)
```
